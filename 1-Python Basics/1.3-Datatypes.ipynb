```python
## Integer Example
age = 42
type(age)  # This will return <class 'int'>
```

```python
## Floating Point Example
height = 5.9
print(height)
type(height)  # This will return <class 'float'>
```

```python
## string Example
name = "Alice"
print(name)
type(name)  # This will return <class 'str'>
```

```python
## Boolean Example
is_student = True
type(is_student)  # This will return <class 'bool'>
```

```python
a = 5
b = 4

type(a==b)  # This will return <class 'bool'> as the result is False and we are comparing two integers using '=='
```

```python
## Common Errors
result= "Hello" + 5  # This will raise a TypeError because you cannot concatenate a string and an integer
```

```python
result="Hello" + str(5)
print(result)
```
