### Standard Library Overview
- Python's standard library is a vast collection of modules and packages that come bundled with python, providing a wide range of functionalities out of the box. Here's an overview of some of the most commonly used modules and packages in the python standard library.

- In the following we cover following standard library:
    - array 
    - math
    - random
    - os
    - shutil
    - json
    - csv
    - datetime
    - time
    - regrex

```python
import array
arr = array.array('i', [1,2,3,4])
print(arr)
```

```python
import math
print(math.sqrt(16))
print(math.pi)
```

```python
# Random 
import random
print(random.randint(1,10))    # this will select a random number between given range
print(random.choice(['apple', 'banana', 'cherry'])) 
```

```python
# File and Directory access
import os
print(os.getcwd())   # getting current working directory path
```

```python
os.mkdir('test_dir')
```

```python
# High level operations on files and collection of files
import shutil
shutil.copyfile('source.txt', 'destination.txt')
```

```python
# Data Serialization
import json
data={'name':'Mani', 'age':32}

json_str=json.dumps(data)
print(json_str)
print(type(json_str))

parsed_data=json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
## CSV 
import csv

with open('example.csv', mode='w', newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name', 'age'])
    writer.writerow({'Mani', 32})

with open('example.csv', mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
# Datetime 
from datetime import datetime, timedelta

now=datetime.now()
print(now)

yesterday=now-timedelta(days=1)
print(yesterday)
```

```python
# Time
import time
print(time.time())

time.sleep(2)
print(time.time())
```

```python
# Regular expression
import re

pattern=r'\d+'
text='There are 123 apples 456'
match=re.search(pattern, text)
print(match.group())
```
