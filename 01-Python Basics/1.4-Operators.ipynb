### 1. Introduction to Operators
- Operators are special symbols or keywords that perform operations on one or more operands (values or variables) and produce a result.

### 2. Arithmetic Operations

```python
a = 10
b = 5

add_result = a + b  # addition
print("Addition:", add_result)

sub_result = a - b  # subtraction
print("Subtraction:", sub_result)

mult_result = a * b # multiplication
print("Multiplication:", mult_result)

div_result = a / b  # division  (gives quotient)
print("Division:", div_result)

mod_result = a % b  # modulus (returns remainder)
print("Modulus:", mod_result)

floor_div_result = a // b  # floor division (returns the largest integer less than or equal to the division result)
print("Floor Division:", floor_div_result)

exponent_result = a ** b  # exponentiation
print("Exponentiation:", exponent_result)
```

### 3. Comparision Operators

```python
## 1. Equal to (==)
a = 10
b = 10
print("a == b:", a == b)
```

```python
str1 = "Hello"
str2 = "Bye"

str1 == str2  # This will return False as both strings are not equal
```

```python
## 2. Not equal to (!=)
a = 10 
b = 5
print("a != b:", a != b)
```

```python
str1 = "Hello"
str2 = "World"
print("str1 != str2:", str1 == str2) # This will return True as both strings are not equal  
```

```python
## 3. Greater than (>)

a = 10
b = 5
print("a > b:", a > b)
```

```python
## 4. Less than (<)

a = 10
b = 5  
print("a < b:", a < b)
```

```python
## 5. Greater than or equal to (>=)

a = 10
b = 5
print("a >= b:", a >= b)
```

```python
## 6. Less than or equal to (<=)
a = 10
b = 5
print("a <= b:", a <= b)
```

### 4. Logical Operators

```python
## And operator
x=True
y=True

result= x and y  # True if both x and y are True
print("x and y:", result)
```

```python
x=False
y=True
 
result= x and y  # True if both x and y are True
print("x and y:", result)
```

```python
## OR operator 
x=False
y=False

result= x or y  # True if either x or y is True
print("x OR y:", result)
```

```python
## NOT operator

x=True
not x  # This will return False as x is True
```

### 5. Simple calculator

```python
# Simple calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Performing arithmetic operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2
floor_division = num1 // num2
modulus = num1 % num2
exponentiation = num1 ** num2

# Displaying results
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
print("Floor Division:", floor_division)
print("Modulus:", modulus)
print("Exponentiation:", exponentiation)
```
