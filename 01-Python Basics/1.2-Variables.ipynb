### Variables
- Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In Python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable.

```python
a = 100
```

```python
## Declaring and Assigning Variables in Python

age = 31
height = 6.0
name = "Mani"
is_student = True

## printing the variables
print("age:", age)
print("height:", height)
print("Name:", name)
```

### Naming Conventions
- Variable names should be descriptive
- They must start with a letter or '_' and contains letter, numbers and underscore
- Variables names are cases sensitive

```python
## Valid variable names
first_name = "RUCHI"
last_name = "RATHOD"
```

```python
## Invalid variable names
# 2age = 25
# first-name = "John"
@name= "Mani"
```

### Case sensitivity
- Python is case sensitive languae

```python
name = "Ruchi"
Name = "Mani"
```

```python
if name == Name:
    print('True')
else:
    print('False')
```

### Understanding variable types
-  Python is dynamically typed, type of a variable is determined at runtime

```python
age = 32 # integer
height = 6.1  # float
name = "Ruchi"  # string
is_student = False  # boolean

print(type(name))
```

```python
## Type checking and conversion in Python
type(age)
```

```python
## type casting / conversion

age = 25
print(type(age))

# Type conversion
age_str = str(age)
print(age_str)
print(type(age_str))
```

### Dynamic Typing
- Python allows the type of a variable to change as the program executes


```python
var=10 #int
print(var, type(var))

var= "Waterfall" 
print(var, type(var))

var=3.14 #float
print(var, type(var))
```

### Simple Calculator

```python
## Lets build a simple calculator that takes two numbers as input and performs addition, subtraction, multiplication, and division.
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("First Number:", num1)
print("Second Number:", num2)
print("Sum:", sum)
print("Difference:", difference)  
print("Product:", product)
print("Quotient:", quotient)
```
