### Decorators:
- Decorators are a powerful and flexible feature in python that allows you to modify the beavior of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code. This lesson covers the basics of decorators, including how ton create and use them.

```python
# Function Copy: A duplicate function that works independently of the original, so changes to one don't affect the other.
def welcome():
    return "Welcome to the advanced python course"

welcome()
```

```python
# Closures Function: A function that remembers variables from its outer scope even after that scope is gone.

def main_welcome(msg):

    def sub_welcome_method():
        print("Welcome to the advance python course")
        print(msg)
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
# Decorator: A function that wraps another function to modify/extend its behavior without changing its code.
def main_welcome(func):
   
    def sub_welcome_method():
        print("Welcome to the advance python course")
        func()
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
def course_introduction():
    print("This is an advanced python course")

coure_introduction()
```

```python
main_welcome(course_introduction)
```

```python
@main_welcome
def coure_introduction():
    print("This is an advanced python course")
```

```python
## Decorator

def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hello!")
```

```python
say_hello()
```

```python
## Decorators WWith arguments
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("Hello")
```

```python
say_hello()
```

#### Conclusion
Decorators are a powerful tool in Python for extending and modifying the behavior of functions and methods. They provide a clean and readable way to add functionality such as logging, timing, access control, and more without changing the original code.
