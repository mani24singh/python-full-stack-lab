### Magic Methods

-  Magic methods in python, also known as dunder methods (double underscore methods), are special methods that start and end with double underscores. These methods enable you to define the behavior of objects for built-in operations, such as arithmetic operations, comparisions and more.

- Magic Methods
    - Magic methods are predefined methods in Python that you can override to change the behavior of your objects. Some common magic methods include:

```python
'''
__init__': Initializes a new instance of a class.
__str__: Returns a string representation of an object.
__repr__: Returns an official string representation of an object.
__len__: Returns the length of an object.
__getitem__: Gets an item from a container.
__setitem__: Sets an item in a container.

'''
```

```python
# Basic Methods
class Person:
    def __init__(self, name, age):
        self.name=name
        self.age=age

    def __str__(self):
        return f"{self.name}, {self.age} years old"
    
    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"

person=Person("Mani", 34)
print(person)
print(repr(person))
```

```python
# Example class using various dunder methods
class Book:
    def __init__(self, title, author, pages, chapters):
        self.title = title
        self.author = author
        self.pages = pages
        self.chapters = chapters  # list of chapter titles
    
    # __str__: For readable string representation
    def __str__(self):
        return f"'{self.title}' by {self.author}"
    
    # __repr__: For debugging/developer representation
    def __repr__(self):
        return f"Book('{self.title}', '{self.author}', {self.pages})"
    
    # __len__: Returns length (pages)
    def __len__(self):
        return self.pages
    
    # __getitem__: Access chapters by index
    def __getitem__(self, index):
        return self.chapters[index]
    
    # __setitem__: Modify chapters by index
    def __setitem__(self, index, value):
        self.chapters[index] = value

# 1. __init__ example
book = Book("1984", "George Orwell", 328, ["The Party", "Big Brother", "Thoughtcrime"])
print("1. __init__: Initializes the object")
print(f"   book.title = {book.title}")
print(f"   book.pages = {book.pages}")
print()

# 2. __str__ example
print("2. __str__: String representation for users")
print(f"   str(book) = {str(book)}")
print(f"   print(book) = ", end="")
print(book)
print()

# 3. __repr__ example
print("3. __repr__: String representation for debugging")
print(f"   repr(book) = {repr(book)}")
print()

# 4. __len__ example
print("4. __len__: Get length of object")
print(f"   len(book) = {len(book)}  # returns number of pages")
print()

# 5. __getitem__ example
print("5. __getitem__: Get item by index")
print(f"   book[0] = {book[0]}  # first chapter")
print(f"   book[1] = {book[1]}  # second chapter")
print()

# 6. __setitem__ example
print("6. __setitem__: Modify item by index")
print(f"   Before: book[2] = {book[2]}")
book[2] = "Doublethink"
print(f"   After: book[2] = {book[2]}")
print()

# Additional demonstration
print("Bonus: Using list operations on chapters")
print(f"   All chapters: {book.chapters}")
print(f"   First two chapters: {book[0:2]}")
```
