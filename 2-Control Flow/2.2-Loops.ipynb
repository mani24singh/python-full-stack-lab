### 1. Introduction to Loops
- Loops are used to execute a block of code repeatedly until a certain condition is met.

### 2. for loop
- The `for` loop is used to iterate over a sequence (like a list, tuple, dictionary, set, or string) or other iterable objects.

```python
for i in range(5):
    print(i)
```

```python
for i in range(1,6):
    print(i)
```

```python
for i in range(1, 10, 2):  # in range function, first parameter is start, second is stop, third is step
    print(i)
```

```python
for i in range(10, 1, -1):  # counting backwards from 10 to 2
    print(i)
```

```python
for i in range(10, 1, -2): # stepping backwards by 2
    print(i)
```

```python
## String Iteration

str = "Mani Singh"

for i in str:
    print(i)
```

### 3. While Loop
- The `while` loop is used to execute a block of code repeatedly as long as a given condition is true.

```python
count = 0

while count < 5:
    print("Count is:", count)
    count += 1
```

### 4. Loop Control statements

1. Break: The break statement exists the loop prematurely

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

2. Continue: The continue statement skips the current iteration and continues with the next

```python
for i in range(12):
    if i % 2 == 0:
        continue
    print(i)
```

3. Pass: The pass statement is a null operation; it does nothing

```python
for i in range(5):
    if i == 3:
        pass  # placeholder for future code
    print(i)
```

### 5. Nested loops
- A loop inside a loop

```python
for i in range(3):    # 0, 1, 2
    for j in range(2):   # 0, 1
        print(f"i: {i}, and j: {j}")
```

### 6. Practical Examples

### Examples - Calculate the sum of first N natural numbers using a a while and for loop


```python
## While loop

n = 10
sum = 0
count = 1

while count <= n:
    sum = sum + count
    count += 1
print("The sum of first", n, "numbers is:", sum)
```

```python
## for loop 
n = 10
sum = 0
for i in range(n+1):
    sum = sum + i
print("The sum of first", n, "numbers is:", sum)
```

### Examples - Prime Numbers betweeen 1 to 100

```python
for n in range(1, 101):
    if n > 1:
        for i in range(2, n):
            if (n%i) == 0:
                break
        else:
            print(n)
```
