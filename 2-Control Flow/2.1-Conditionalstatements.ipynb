### 1. Introduction to conditional statements
- Conditional statements allow you to execute specific blocks of code based on certain conditions. 
- They help in making decisions within your code, enabling different outcomes based on varying inputs or states.

### 2. If statement

```python
age=40

if age>=18:
    print("Valid age for voting")
```

```python
age>=18
```

### 3. Else statement
- The else statement executes a block of code if the condition in the if statement is false

```python
age=16

if age>=18:
    print("You are eligible for voting")
else:
    print("You are a minor")
```

### 4. Elif statement
- The elif statement allows to check multiple conditions. Its stands for "else if"

```python
age=40

if age >= 18:
    print("You are an adult.")
elif age >= 13:
    print("You are a teenager.")
else:
    print("You are a child.")
```

### 5. Nested Conditional Statements

- We can create one or more if, elif or else statement inside another if, elif or else statement to create nested conditional statements.

```python
### Number even, odd, negative

num = int(input("Enter a number: "))

if num > 0:
    print("Number is positive.")
    if num%2 == 0:
        print("Number is a positive even number.")
    elif num%2 != 0:
        print("Number is a positive odd number.")
else:
    print("Number is zero or negative.")
```

### 6. Practical Examples:

- Determine if a year is a leap year using nested condition statement

    - A year is a leap year if:
        - It is divisible by 4
        - Except years that are divisible by 100
        - Unless the year is also divisible by 400

    - In simple words
        - Every 4th year → leap year
        - But every 100th year → not a leap year
        - But every 400th year → leap year again

```python
year = int(input("Enter a year: "))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("The year", year, "is a leap year.")
        else:
            print("The year", year, "is not a leap year.")
    else:
        print("The year", year, "is a leap year.")
else:
    print("The year", year, "is not a leap year.")
```

### Assignments

- Write a program to create a simple calculator where we take two numbers as input from user

```python
## take user input num1, num2
num1 = float(input("Enter first  number:"))
num2 = float(input("Enter second number:"))
operation = input("Enter operation (+, -, *, /): ")

# Perform operation based on user input
if operation == '+':
    result = num1 + num2
elif operation == '-':
    if num1 >= num2:
        result = num1 - num2
    else:
        result = num2 - num1
elif operation == '*':
    result = num1 * num2
elif operation == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error: Division by zero is not allowed."
else:
    result = "Error: Invalid operation."

print("The result is:", result)
```

- Determine the ticket price based on age and whether the person is a student

```python
## Ticket pricing based on age and student status

# Take user input for age and student status
age = int(input("Enter your age:"))
is_student = input("Are you a student? (yes/no):").strip().lower()

# Determine ticket price
if age < 5:
    price = "Free"
elif age <= 12:
    price = "$10"
elif age <= 17:
    if is_student == 'yes':
        price = "$12"
    else:
        price = "$15"
elif age <= 64:
    if is_student == 'yes':
        price = "$18"
    else:
        price = "$25"
else:
    price = "$20"

print("Your ticket price is:", price)
```

### Complex Example: Employee Bonus Calculation

- Calculate an Employee's bonus based on their performance rating and years of service

```python
# Employee bonus calculation

# Take user input for years of service and performance rating 
years_of_service = int(input("Enter years of service:"))
peformance_rating = float(input("Enter performance rating (1.0 to 5.0)"))

# Determine the bonus percentage
if peformance_rating >= 4.5:
    if years_of_service > 10:
        bonus_percentage = 20
    elif years_of_service >= 5:
        bonus_percentage = 15
    else:
        bonus_percentage = 10
elif peformance_rating >= 3.5:
    if years_of_service > 10:
        bonus_percentage = 15
    elif years_of_service >= 5:
        bonus_percentage = 10
    else:
        bonus_percentage = 5
else: 
    bonus_percentage = 0

# Calculate bonus amount 
salary = float(input("Enter your salary:"))
bonus_amount = (bonus_percentage / 100) * salary

print("Years of service:", years_of_service)
print("Performance rating:", peformance_rating)
print("Bonus percentage is:", bonus_percentage, "%")
print("Bonus amount is: ${:.2f}".format(bonus_amount))
```

### Complex Example: User Login System

- A simple user login system that checks the username and password

```python
## user login system

# Predefined username and password
stored_username = "admin"
stored_password = "password123"

# Take user input for username and password
username = input("Enter username:")
password = input("Enter password:")

# Check the login credentials
if username == stored_username:
    if password == stored_password:
        print("Login successful! Welcome,", username)
    else:
        print("Incorrect password. Please try again.")
else:
    print("Username not found. Please try again.")
```
