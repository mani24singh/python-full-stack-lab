### Real World Examples using Lists in Python:
- Lists are one of the most commonly used data structures in Python, due to its versatility and ease of use.  Here are several practical examples that illustratre their use in real-world scenarios.

## Example 1: Manage a TO-DO List
- Create a TO-DO List to keep track of tasks.

```python
to_do_list = ["buy groceries", "clean the house", "pay bills", "read a book", "Finish assignment", "Visit Bank"]
print(to_do_list)

## Adding a task
to_do_list.append("Schedule Meeting")
to_do_list.append("Go to Gym")

## Removing a completed task
to_do_list.remove("clean the house")

## Check if a task is in the list
if "pay bills" in to_do_list:
    print("Task found: pay bills")
print("To Do List Remaining")
for task in to_do_list:
    print(f"-{task}")
```

### Example 2: Organizing student grades
- Create a list to store and calculate average grades for students

```python
# organizing student grades
grades = [88, 92, 79, 85, 90, 95, 78, 84]

# adding a new grade
grades.append(91)

# calculating average grade
average_grade = sum(grades)/len(grades)
print(f"Average Grade: {average_grade:.2f}")

# Findng the highest and lowest grades
highest_grade = max(grades)
lowest_grade = min(grades)
print(f"Highest Grade: {highest_grade}")
print(f"Lowest Grade: {lowest_grade}")
```

### Example 3: Managing an Inventory
- Use a list to manage inventory items in a store

```python
# Managing a list of inventory items (fruits)
inventory = ["apple", "banana", "orange", "mango", "grape"]
print("Initial Inventory:", inventory)

# Adding a new item
inventory.append("kiwi")
print("Updated Inventory:", inventory)

# Removing an item that is out of stock
inventory.remove("banana")
print("Inventory after removing banana:", inventory)

# Checking if an item is out of stock
item = "oranges"
if item in inventory:
    print(f"{item} is in stock.")
else:
    print(f"{item} is out of stock.")

# Printing all items in the inventory
print("Inventory Items:")
for item in inventory:
    print(f"- {item}")
```

### Example 4: Collecting User  Feedback
- Use a list to collect and analyze user feedback

```python
# Collecting user feedback
feedback = ["Great Service!", "very satisfied", "Could be better", "Excellent quality", "Not happy with the delivery"]
print(feedback)

# Adding new feedback
feedback.append("Not happy with the service!")

# Counting specific feedback
positive_feedback_count = sum([1 for comment in feedback if "great" in comment.lower() or "excellent" in comment.lower() or "satisfied" in comment.lower()])
print(f"Number of positive feedbacks: {positive_feedback_count}")

# Printing all feedback comments
print("User Feedback:")
for comment in feedback:
    print(f"- {comment}")
```
