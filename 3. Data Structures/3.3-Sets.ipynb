### 1. Introduction to Sets

- Sets are built-in data type in python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific ordder, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.

### 1. Creating a Set

```python
# Create a set
my_set = {1,2,3,4,5}
print(my_set)

print(type(my_set))
```

```python
my_empty_set = set()
print(type(my_empty_set))
```

```python
my_set = set([1,2,3,4,5,6]) # convert list to set
print(my_set)
```

### 3. Basic Set Operations

```python
# Basic set operations

# 1. Adding and removing elements
my_set.add(7)
print("After adding 7:", my_set)

my_set.remove(3)
print("After removing 3:", my_set)
```

```python
my_set.remove(10)  # This will raise a KeyError since 10 is not in the set
```

```python
my_set.discard(10) # This will not raise an error even if 10 is not in the set
print(my_set)
```

```python
# Pop method - removes and returns an arbitrary element from the set
removed_element = my_set.pop()
print("Removed element:", removed_element)
print("Set after pop:", my_set)
```

```python
# Clear all the elements from the set
my_set.clear()
print("Set after clear:", my_set)
```

### 4. Membership Testing for set

```python
# Set membership testing
my_set = {1,2,3,4,5}
print(3 in my_set)  # True
print(10 in my_set) # False
```

### 5. Mathematical set operations

```python
# Mathematical set operations
set1 = {1,2,3,4}
set2 = {3,4,5,6}

# Union: Union of two sets is a set containing all elements from both sets
set_union = set1.union(set2) 
print("Union:", set_union)

# Intersection: Intersection of two sets is a set containing only elements that are in both sets
set_intersection = set1.intersection(set2)
print("Intersection:", set_intersection)

# In-place intersection update: means updating the first set to keep only elements found in both sets
set1.intersection_update(set2)
print("After intersection update, set1:", set1)
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9} 

# Difference: Difference of two sets is a set containing elements that are in the first set but not in the second set
set_difference = set1.difference(set2)
print("Difference:", set_difference)
```

```python
set1
```

```python
set2.difference(set1)
```

```python
# Difference update: means updating the first set to remove elements found in the second set
set1.difference_update(set2)
print("After difference update, set1:", set1)
```

```python
set1 = {1,2,3,4,5}
set2 = {4,5,6,7,8}

# Symmetric Difference: Symmetric difference of two sets is a set containing elements that are in either of the sets but not in both
set_sym_diff = set1.symmetric_difference(set2)
print("Symmetric Difference:", set_sym_diff)
```

### 6. Set Methods

```python
# Set Methods
set1 = {1,2,3,4,5}
set2 = {3,4,5}

# issubset: checks if set2 is a subset of set1 which means all elements of set2 are in set1
print(set1.issubset(set2))  # True

# issuperset: checks if set1 is a superset of set2 which means all elements of set2 are in set1
print(set1.issuperset(set2))  # True
```

```python
lst = [1,2,2,3,4,4,5]
set(lst)  # Removes duplicates
```

### 7. Practical Example for set

```python
# Counting unique words in text

text = "In this tutorial we will learn about sets in Python. Sets are useful for storing unique items."
words = text.lower().split()
print("Words:", words)
print("Total words:", len(words))

# Convert list of words to a set to get unique words
unique_words = set(words)
print("Unique words:", unique_words)
print("Number of unique words:", len(unique_words))
```
