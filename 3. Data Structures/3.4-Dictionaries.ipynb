### 1. Introduction to Dictionaries

- Dictionaries are unordered collection of item. They store data in key-value pairs. Keys must be unique and immutable (eg. strings, numbers, or tuples), while values can be of any type.


### 2. Creating Dictionaries

```python
# Creating Dictionaries
empty_dict = {}
print(type(empty_dict)) 
```

```python
empty_dict = dict()
print(empty_dict)
print(type(empty_dict))
```

```python
student = {"name": "Ruchi", "age":28, "grade": 93}
print(student)

print(type(student))
```

```python
# Single key is always used only once in a dictionary
student = {"name": "Ruchi", "age":28, "name": "Mani"}
print(student) 
```

### 3. Accesssing Dictionary elements

```python
# Accessing Dictionary elements
student = {"name": "Ruchi", "age":28, "grade": 93}
print(student)
```

```python
# Accessing Dictionary elements (access values using keys)
print(student["name"])  
print(student["age"])

# Alternate way: Accessing using get() method
print(student.get("age"))  
print(student.get("grade"))

print(student.get("address"))  # None
print(student.get("address", "Not Available"))  # Not Available means default value if key not found
```

### 4. Modifying dictionary elements
- Dictionaries are mutable, so we can update existing key-value pairs or add new ones or delete them.

```python
print("Before modification:", student)

student["age"] = 23 # Update existing key
print("After updating age:", student)

student["address"] = "123 Main St" # Add new key-value pair
print("After adding address:", student)
```

```python
# deleting key-value pairs
del student["grade"]
print("After deleting grade:", student)
```

### 5. Dictionary methods
- Dictionaries come with several built-in methods for common operations like adding, removing, and accessing elements.

```python
keys = student.keys() # Get all keys
print("Keys:", keys)

values = student.values() # Get all values
print("Values:", values)

items = student.items() # Get all key-value pairs
print("Items:", items)
```

### 6. Shallow Copy vs Deep Copy of Dictionaries
- Shallow Copy: A shallow copy of a dictionary creates a new dictionary object, but does not create copies of nested objects within the original dictionary. Instead, it copies references to those objects. Therefore, changes made to nested objects in the copied dictionary will also affect the original dictionary.

- Deep Copy: A deep copy of a dictionary creates a new dictionary object and recursively copies all nested objects within the original dictionary. This means that changes made to nested objects in the copied dictionary will not affect the original dictionary, as they are completely independent of each other.

```python
# Shallow Copy
original = {"name": "Ruchi", "age":28, "grade": 93}
print("Original Dictionary:", original)

shallow_copy = original  # Both variables point to the same dictionary
print("Shallow Copy Dictionary:", shallow_copy)
```

```python
original["age"] = 29  # Update existing key
print("After modifying original dictionary:", original)
print("Shallow Copy reflects changes:", shallow_copy)
```

```python
# Deep Copy
original = {"name": "Ruchi", "age":28, "grade": 93}
print("Original Dictionary:", original)

deep_copy = original.copy()  # Creates a new dictionary with the same content
print("Deep Copy Dictionary:", deep_copy)
```

```python
original["age"] = 30  # Update existing key
print("After modifying original dictionary:", original)
print("Deep Copy remains unchanged:", deep_copy)
```

### 7. Iterating over dictionaries
- We can use loops to iterate over dictionaries (keys, values, or items).

```python
# iterating over keys

student = {"name": "Ruchi", "age":28, "grade": 93}

for keys in student.keys():
    print(keys)
```

```python
# iterating over values
for values in student.values():
    print(values)   
```

```python
# iterating over key value pairs
for key, value in student.items()   :
    print(f"{key}: {value}")
```

### 8. Nested Dictionaries
- A nested dictionary is a dictionary that contains another dictionary as a value. This allows for more complex data structures.

```python
students = {
    "student1": {"name": "Ruchi", "age":28, "grade": 93},
    "student2": {"name": "Mani", "age":22, "grade": 88}
}

print(students)
```

```python
# Accessing nested dictionary elements
print(students["student1"]["name"])
print(students["student2"]["grade"])
```

```python
students.items()
```

```python
# iterating over nested dictionaries
for student_id, student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key, value in student_info.items():
        print(f"{key}: {value}")
```

### 9. Dictionary Comprehension
- Dictionary comprehensions provide a concise way to create dictionaries. They consist of an expression pair (key: value) followed by a for clause, and can also include optional if clauses.

- syntax: 
    - {key_expression: value_expression for item in iterable if condition}

```python
squares = {x:x*2 for x in range(5)}
print(squares) 
```

```python
# Conditional dictionary comprehension  
even_squares = {x:x*2 for x in range(10) if x%2==0}
print(even_squares)
```

### 10. Practical Examples

```python
# 1. Use a dictionary to count the frequency of each element in a list.

number = {1,2,2,3,4,4,4,5}
frequency = {}

for num in number:
    if num in frequency:
        frequency[num] +=1
    else:
        frequency[num] =1
print(frequency)
```

```python
# Merge 2 dictionaries into one.

dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}

merged_dict = {**dict1, **dict2}  # dict2 values will overwrite dict1 for duplicate keys, here ** means unpacking
print(merged_dict) 
```
