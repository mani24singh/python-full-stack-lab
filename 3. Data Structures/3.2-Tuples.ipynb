### 1. Introduction to Tuples
- Explaination: Tuples are ordered collections of items that are immutable. They are similar to lists, but their immutability makes them different.

### 2. Creating Tuples

```python
# Creating a tuple
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(type(lst))

tpl = tuple()
print(type(tpl))
```

```python
numbers = tuple([1, 2, 3, 4, 5])
print(numbers)
```

```python
list((1,2,3,4,5))
```

Hence, the code above demonstrates how to create and manipulate tuples in Python, including creating an empty tuple, converting a list to a tuple, and converting a tuple back to a list.

### 5. Immutable nature of tuples

```python
## Tuples are immutable, meaning their elements cannot be changed once assigned.

lst = [1,2,3,4,5]
print("Original List:", lst)

lst[1] = "Mani"
print("Modified List:", lst)

## Hence, this shows that list is mutable.
```

```python
## Whereas, the tuple is immutable.
numbers[1] = "Mani"  # This will raise a TypeError
```

```python
numbers
```

### 6. Tuple Methods

```python
# Tuples Methods
print(numbers.count(3))
print(numbers.index(4))
```

### 7. Packing and Unpacking of Tuples

```python
# 1. Packing:
packed_tuple = 1, "Mani", 3.14, False
print(packed_tuple)
```

```python
# 2. Unpacking:
a, b, c, d = packed_tuple

print(a)
print(b)    
print(c)
```

```python
# Unpacking a tuple with *
numbers = (1, 2, 3, 4, 5, 6)
first, *middle, last = numbers

print(first)
print(middle)
print(last)
```

### 8. Nested Tuple

```python
# here is a list inside a list (nested list)
lst = [[1,2,3], [4,5,6], ["Mani", "is", "here"]]
lst[0][0:3]
```

```python
# here is a tuple inside a list
lst = [[1,2,3], [4,5,6], ("This", "is", "a", "tuple")] 
lst[2][0:4]
```

```python
# Nested tuple
nested_tuple = ((1,2,3), ("a", "b", "c"),  (True, False), 7, "Python")
print(nested_tuple)
```

```python
# accessing the elements of nested tuple
print(nested_tuple[0])        # First inner tuple
print(nested_tuple[1][1])     # Second element of second inner tuple
print(nested_tuple[2][0])     # First element of third inner tuple
```

```python
# Iterating through a nested tuple
tuple_nest = ((1, 2, 3), ('a', 'b', 'c'), (True, False))
for sub_tuple in tuple_nest:
    for item in sub_tuple:
        print(item, end=' ') # here end=' ' is used to print in the same line
    print()
```

### 9. Practical Examples and Common Errors

```python
# Function that returns multiple values using tuple packing and unpacking
def calculate(a, b):
    sum_ab = a + b
    diff_ab = a - b
    prod_ab = a * b
    return sum_ab, diff_ab, prod_ab  # packed as a tuple

# Calling the function and unpacking the returned tuple
result = calculate(10, 5)
sum_result, diff_result, prod_result = result  # unpacking  
print("Sum:", sum_result)
print("Difference:", diff_result)
print("Product:", prod_result)

# Alternatively, unpacking directly during the function call
sum_result, diff_result, prod_result = calculate(10, 5)
print("Sum:", sum_result)
print("Difference:", diff_result)
print("Product:", prod_result)
```

```python
# Common Errors - we have already covered those above, just summarizing here agin

# 1. Mismatched Unpacking
packed_data = (1, 2, 3)
# a, b = packed_data  # This will raise a ValueError

# 2. Attempting to Modify a Tuple
immutable_tuple = (1, 2, 3)
# immutable_tuple[0] = 10  # This will raise a TypeError

# 3. Incorrect Indexing
sample_tuple = (1, 2, 3)
# print(sample_tuple[3])  # This will raise an IndexError
```
