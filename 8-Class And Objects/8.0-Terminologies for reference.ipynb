### OOPS ‚Äì Quick Terminology Memory (for Classes in Python)

1. `def __init__(self, ...)`  
   ‚Üí **constructor**  
   ‚Üí runs automatically when an object is created

2. `def selfdriving(self):`  
   ‚Üí **instance method**  
   ‚Üí works on object data using `self`

3. Any `def` written directly inside a class  
   ‚Üí **method** (generic term)

4. `@classmethod` + `def method(cls, ...)`  
   ‚Üí **class method**  
   ‚Üí works on class-level data using `cls`

5. `@staticmethod` + `def method(...)`  
   ‚Üí **static method**  
   ‚Üí utility/helper function related to the class

6. `@property` + `def value(self):`  
   ‚Üí **property** (getter)  
   ‚Üí accessed like an attribute: `obj.value`

7. `@value.setter`  
   ‚Üí **setter method**  
   ‚Üí controls how a property is updated: `obj.value = 10`

8. `self.x = 10` inside `__init__`  
   ‚Üí **instance variable / attribute**  
   ‚Üí belongs to each object

9. `class Car: wheels = 4`  
   ‚Üí **class variable / attribute**  
   ‚Üí shared by all objects

10. `class ElectricCar(Car):`  
    ‚Üí **inheritance**  
    ‚Üí `ElectricCar` is a *child/subclass* of `Car`

11. `super().__init__(...)`  
    ‚Üí calls the **parent class constructor**

12. `def __str__(self):`  
    ‚Üí **string representation method**  
    ‚Üí used by `print(obj)`

13. `def __repr__(self):`  
    ‚Üí **developer/debug representation**

14. `def __len__(self):`  
    ‚Üí enables `len(obj)`

15. `def __eq__(self, other):`  
    ‚Üí defines `obj1 == obj2`

16. `def __add__(self, other):`  
    ‚Üí defines `obj1 + obj2`

17. `self._value`  
    ‚Üí **protected attribute** (by convention)

18. `self.__value`  
    ‚Üí **private attribute** (name-mangled)

19. `isinstance(obj, ClassName)`  
    ‚Üí **type check** for an object

20. `hasattr(obj, "attr")`  
    ‚Üí checks if an object **has an attribute**



Think of it like a mental map:

- `self`  ‚Üí object-level  
- `cls`   ‚Üí class-level  
- `@`     ‚Üí behavior modifier (decorator)  
- `__xx__` ‚Üí Python‚Äôs internal ‚Äúmagic hooks‚Äù  
- `_x`    ‚Üí ‚Äúplease don‚Äôt touch‚Äù  
- `__x`   ‚Üí ‚Äútrying to hide‚Äù
