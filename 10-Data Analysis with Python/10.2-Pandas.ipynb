### Pandas DataFrame and Series

- Pandas is a powerful data manipulation library in python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. 
- A Series is a one-dimensional array-like object
- While a DataFrame is a two-dimensional, size-mutable and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

```python
# pip install pandas
```

```python
import pandas as pd
```

- Series: A pandas series is a one-dimensional array like object that can hold any data type. It is similar to a column in a table.

```python
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series \n", series)
print(type(series))
```

```python
# Create a series from dictionary
data={'a':1, 'b':2, 'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data, index=index)
```

- DataFrame: 

```python
# create a DataFrame from a dictionary of list

data ={
    'Name': ['Mani', 'Ruchi', 'Dolly'],
    'Age': [31, 29, 26],
    'City': ['Mumbai', 'Manchester', 'Florida']
}

df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
# Create a DataFrame from a list of dictionaries
data = [
    {'Name': 'Mani', 'Age':32, 'City':'Mumbai'},
    {'Name': 'Ruchi', 'Age':29, 'City':'Manchester'},
    {'Name': 'Dolly', 'Age':26, 'City':'Pune'},
    {'Name': 'Lucky', 'Age':29, 'City':'Nashik'},
]

df=pd.DataFrame(data)
print(df)
print(type(df))
```
